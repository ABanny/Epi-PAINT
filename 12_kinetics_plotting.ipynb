```python
import colorsys
import numpy as np

n_channels = 12
hues = np.arange(0, 1, 1 / n_channels)
colors = [colorsys.hsv_to_rgb(_, 1, 1) for _ in hues]
colors = colors[6:] + colors[:6]
```

```python
grid = sns.FacetGrid(data, col="Imager", hue="Imager", palette=colors,
                     col_wrap=6, height=3)

grid.map(sns.scatterplot, "dark_mean", "len_mean", alpha = 0.1)

# grid.figure.subplots_adjust(wspace=0)

n_rows = 2

# Iterate through all axes
for i, ax in enumerate(grid.axes.flat):
    # Check if the current subplot is NOT the first in its row
    if i < 6:  # Skip the first plot in each row
        ax.xaxis.set_visible(False)  # Hide x-axis ticks and labels
        # ax.spines['left'].set_visible(False)  # Hide spine

grid.set_axis_labels(x_var='Dark Time (s)', y_var='Bright Time (s)')

plt.xlim(1,10000)
plt.ylim(0.1, 5)
plt.yscale('log')
plt.xscale('log')
# plt.xlabel('Dark Time')

plt.savefig("/Users/abhinav/Library/CloudStorage/OneDrive-IndianInstituteofScience/Papers/zz_Msequences/Figures/Kinetics/supp_1.pdf", format="pdf")
```
